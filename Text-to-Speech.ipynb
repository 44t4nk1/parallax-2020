```
import speech_recognition as sr
import time
rec=sr.Recognizer()

audio='/content/gdrive/My Drive/Colab Notebooks/the_audio.wav'

with sr.AudioFile(audio) as source:
  #rec.adjust_for_(source)
  audio_1=rec.record(source,duration=6)
  audio_2=rec.record(source,duration=6)
  audio_3=rec.record(source,duration=6)
  audio_4=rec.record(source,duration=6)
text=rec.recognize_google(audio_1)+rec.recognize_google(audio_2)+rec.recognize_google(audio_3)+rec.recognize_google(audio_4)

```

```
text
```

```
from googletrans import Translator
T=Translator()
text_hindi=T.translate(text,src='en',dest='hi').text
text_tamil=T.translate(text,src='en',dest='ta').text
```
