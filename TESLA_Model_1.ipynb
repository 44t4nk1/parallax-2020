```
from datetime import datetime as dt,timedelta as td
import pandas_datareader.data as web
import pandas as pd
import numpy as np
```

```
start_date=dt(2010,1,1)
end=dt.strftime(dt.today()-td(1),'%Y-%m-%d')
df_sbi=web.DataReader('TSLA','yahoo',start_date,end)#State bank of india
df_sbi.tail()
```

```
X=df_sbi.filter(['Close'])
X=X.values
X
```

```
from sklearn.preprocessing import MinMaxScaler as mms
scaler=mms(feature_range=(0,1))
X=scaler.fit_transform(X)
```

```
import math as m
x_train_len=m.ceil(X.shape[0]*0.8)
x_train_len
```

```
x_train=list()
y_train=list()
for i in range(1,x_train_len):
  x_train.append(X[i-1:i])
  y_train.append(X[i,])
x_train=np.array(x_train)
y_train=np.array(y_train)
x_train=np.reshape(x_train,(x_train.shape[0],x_train.shape[1],1))
x_train.shape
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense

model=Sequential()
model.add(LSTM(70,return_sequences=True,input_shape=(x_train.shape[1],1)))
model.add(LSTM(70,return_sequences=False))
model.add(Dense(70))
model.add(Dense(1))

model.compile(loss='mean_squared_error',optimizer='adam')
model.summary()
```

```
model.fit(x_train,y_train,batch_size=32,epochs=50)
```

```
x_test_len=m.ceil(X.shape[0]*0.2)
X_Test=X[x_test_len:]
x_test=list()
y_test=list()
X_Test=X[x_test_len:]
x_test=list()
y_test=list()
for i in range(1,x_test_len):
  x_test.append(X_Test[i-1:i])
  y_test.append(X_Test[i,])
x_test=np.array(x_test)
y_test=np.array(y_test)
x_test=np.reshape(x_test,(x_test.shape[0],x_test.shape[1],1))
x_test.shape
```

```
(1-model.evaluate(x_test,y_test))*100
```

```
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler as mms
from tensorflow.keras.models import load_model
model = load_model('/content/gdrive/My Drive/Colab Notebooks/sbi_model_jo_checkkkk.h5')
def sbi_result():
  today=dt.strftime(dt.today(),'%Y-%m-%d')
  df_sbi=web.DataReader('TSLA','yahoo',start='2019-1-1',end=today)
  df_sbi=df_sbi.filter(['Close']).values
  scaler=mms(feature_range=(0,1))
  df_sbi=scaler.fit_transform(df_sbi)
  df_sbi_30=df_sbi[-1:,:]
  y_30=list()
  for i in range(30):
    x_30=list()
    df_sbi_30=np.reshape(df_sbi_30,(df_sbi_30.shape[0],1))
  
    x_30=df_sbi_30[-1:,:]
    x_30=np.array(x_30)
    x_30=np.reshape(x_30,(x_30.shape[1],1,1))
    x=model.predict(x_30)
    y_30.append(x)
    df_sbi_30=np.append(df_sbi_30,x)
  y_30=np.array(y_30)
  y_30=np.reshape(y_30,(y_30.shape[0],y_30.shape[1]))
  y_30=scaler.inverse_transform(y_30)
  result=np.reshape(y_30,(30))
  return result
```
